# Linear Regression 

In this regression task we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variables.

**AUTHOR : SHIVAM SHARMA**

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Reading dataset from url
url = "http://bit.ly/w-data"
dataset = pd.read_csv(url)
print("DataSetImported Success ")
```

```python
#printing few values
dataset.head(10)
```

```python
#plotting the data set on graph
dataset.plot(x = "Hours", y = "Scores", style = 'o')
plt.title('Hours vs Percentage hours')
plt.xlabel('Hours')
plt.ylabel('Percentage Hours')
plt.show()
```

DATA PREPROCESSING

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
print(y)
```

SPLITING DATASET INTO TRAIN AND TEST SET

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state=0)
```

```python
print("X_train")
X_train
```

```python
print("X_test")
X_test
```

```python
print("y_train")
y_train
```

```python
print("y_test")
y_test
```

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
Regressor = LinearRegression()
Regressor.fit(X_train,y_train)
print("Model Fitted")
```

```python
# Plotting the regression line
line = Regressor.coef_ * X + Regressor.intercept_

# Plotting for the test data
plt.scatter(X,y)
plt.plot(X, line)
plt.show()
```

Predicting Value

```python
y_pred = Regressor.predict(X_test)
print("Value Predicted")
```

```python
# Comparision Of Actual And Predicting Value
DF = pd.DataFrame({'Actual': y_test , 'Predicted': y_pred})
DF
 
```
